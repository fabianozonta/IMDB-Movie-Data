Classify movies based on ratings

Apply( ) Functions


```python
def rating_group(rating):
    if rating >= 7.5:
        return 'Good'
    elif rating >= 6.0:
        return 'Average'
    else:
        return 'Bad'
```

```python
data['Nova Categoria Rating'] = data['Rating'].apply(rating_group)
```
